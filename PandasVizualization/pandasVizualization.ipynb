```python
#1
#Build corr heatmap
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
selected_columns = ['PTS', 'AST', 'TRB', 'STL', 'BLK', 'TOV', 'FG%', '3P%', 'FT%']
selected_df = Seasons_Stats[selected_columns].dropna()
corr_matrix = selected_df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, fmt=".2f")
plt.title("Heatmap")
plt.show()
```

Thanks to this heatmap, we can analyze the relationships between different statistical metrics. If there is a positive correlation that is close to one, it means that these variables have a strong connection. For example, the number of points scored is related to STL (steals) and TOV (turnovers). This means that as the number of points scored increases, so does the number of steals, and the player tends to turn the ball over more often.
A negative correlation between 3P% (three-point shooting percentage) and the number of rebounds suggests that players who make more three-point shots tend to get fewer rebounds.

```python
#2
import folium
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="geoapiExercises")
df = players.dropna(subset=['birth_city', 'birth_state'])
def get_location(row):
    try:
        location = geolocator.geocode(f"{row['birth_city']}, {row['birth_state']}")
        if location:
            return location.latitude, location.longitude
    except:
        return None, None

df['coords'] = df.apply(get_location, axis=1)

df
```

```python
map = folium.Map(location=[37.0902, -95.7129], zoom_start=5)

for index, row in df.iterrows():
    if row['coords']:
        latitude, longitude = row['coords']
        if latitude is not None and longitude is not None:
            folium.Marker([latitude, longitude], popup=f"{row['birth_city']}, {row['birth_state']}").add_to(map)
map.save('nba_players_birthplaces.html')
```

![Players Birth Map](birthmap.png)
First, we had to transform our birthplaces into coordinates. For this task, we used a geocoder. Afterward, we added a row with coordinates to our table. The variable 'map' starts with the territory of the USA. Thanks to the Folium library, we compiled our results into an HTML file, which contains a map with special markers.

```python
#6
player_data['start_date'] = pd.to_datetime(player_data['year_start'], format='%Y')
player_data['end_date'] = pd.to_datetime(player_data['year_end'], format='%Y')
player_data['career_length'] = (player_data['end_date'] - player_data['start_date']).dt.days


sample_players = player_data.sort_values(by='career_length', ascending=False).head(10)
sample_players = sample_players.reset_index()
plt.figure(figsize=(10, 8))
plt.barh(sample_players.index, sample_players['career_length'], color='skyblue')
plt.xlabel('Lengt of career(in days)')
plt.yticks(sample_players.index, sample_players['name'])
plt.title('First 20 players who had the longest career')
plt.gca().invert_yaxis()
plt.show()
```

This visualization shows the duration of the careers for the players with the longest tenure in the league.

```python
#9
import squarify
players.dropna(subset=['collage'], inplace=True)

college_counts = players['collage'].value_counts()

top_colleges = college_counts.head(10)

others_count = college_counts[10:].sum()
others_series = pd.Series(others_count, index=['Others'])
top_colleges = pd.concat([top_colleges, others_series])

plt.figure(figsize=(12, 8))
squarify.plot(sizes=top_colleges.values, label=top_colleges.index, alpha=0.8, text_kwargs={'fontsize':9})
plt.title('Top Colleges Distribution of NBA Players')
plt.axis('off')
plt.show()
```

This type of chart is useful for quickly understanding which colleges are the main "suppliers" of talent to the NBA.

```python
#10
from math import pi
positions = ['PG', 'SG', 'SF', 'PF', 'C']
position_stats = {}
for position in positions:
    position_data = Seasons_Stats[Seasons_Stats['Pos'] == position]
    position_stats[position] = {
        'PTS': position_data['PTS'].mean(),
        'AST': position_data['AST'].mean(),
        'TRB': position_data['TRB'].mean(),
        'STL': position_data['STL'].mean(),
        'BLK': position_data['BLK'].mean()
    }

categories = list(position_stats[next(iter(position_stats))].keys())
N = len(categories)

angles = [n / float(N) * 2 * pi for n in range(N)]
angles += angles[:1]

fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))

for position, stats in position_stats.items():
    values = list(stats.values())
    values += values[:1]
    ax.plot(angles, values, linewidth=1, linestyle='solid', label=position)

plt.xticks(angles[:-1], categories)
plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))

plt.show()
```


The radar chart allows us to make the following observations:

Point Guards (PG) and Shooting Guards (SG) typically score more points (PTS) and have more assists (AST) compared to other positions.

Centers (C) and Power Forwards (PF) stand out with a higher number of rebounds (TRB) and block shots (BLK), reflecting their role in defending under their own basket and competing for rebounds.

Small Forwards (SF) occupy an intermediate position between the outer and inner lines, showing balanced indicators across all categories.
