```python
from urllib.request import urlopen
import numpy as np

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
our_data = np.genfromtxt(url, delimiter=",", usecols=(0,1,2,3), dtype=float)
our_data
```

```python
#3
print("Our mean->", np.mean(our_data[:, 0]))
print("Our median->", np.median(our_data[:, 0]))
print("Our std->", np.std(our_data[:, 0]))

```

```python
#4
flat_indices = np.arange(our_data.size)
np.random.shuffle(flat_indices)
flat_indices = flat_indices[:20]
row_indices, col_indices = np.unravel_index(flat_indices, our_data.shape)
our_data[row_indices, col_indices] = np.nan
nan_count = np.count_nonzero(np.isnan(our_data))
our_data, nan_count


```

```python
#6
filtered_list = []
filtered_data = our_data[(our_data[:, 2] > 1.5) & (our_data[:, 0] < 5)]
filtered_data
```

```python
#7
filtered_data = np.nan_to_num(filtered_data)
filtered_data
```

```python
#8
print(np.unique(filtered_data, return_counts=True))
```

```python
#9
first_part = filtered_data[:len(filtered_data)//2]
second_part = filtered_data[len(filtered_data)//2:]
first_part, second_part
```

```python
#10
first_part = first_part[first_part[:, 0].argsort()]
second_part = second_part[second_part[:, 0].argsort()[::-1]]
first_part, second_part
```

```python
#11
united_data = np.vstack((first_part, second_part))
united_data
```

```python
#12
flatten_data = united_data.flatten()
unique_numbers, counts = np.unique(flatten_data, return_counts=True)
unique_numbers[np.argmax(counts)]
```

```python
#13,14
def mult_below_adjusted_mean(col):
    mean_value = np.mean(col)
    less_than_mean = col < mean_value
    col[less_than_mean] *= 2
    col[~less_than_mean] /= 4


mult_below_adjusted_mean(united_data[:, 2])
united_data
```
