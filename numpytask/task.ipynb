```python
#1
import re
import urllib.request

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"


with urllib.request.urlopen(url) as response:
    iris_data = re.sub(r'[^\d.,\n]', '', response.read().decode('utf-8')).strip()

print(iris_data)
```

```python
#2
import numpy as np
iris_list = iris_data.split("\n")
data_list = [list(map(float, line[:len(line)-1].split(","))) for line in iris_list]
our_data = np.array(data_list)
print(our_data)
```

```python
#3
print("Our mean->", np.mean(our_data[:, 0]))
print("Our median->", np.median(our_data[:, 0]))
print("Our std->", np.std(our_data[:, 0]))

```

```python
#4
coordinates = []
rows, cols = our_data.shape
for i in range(20):
     while True:
        random_cords = (np.random.randint(rows), np.random.randint(cols))
        if random_cords not in coordinates:
            coordinates.append(random_cords)
            break
for pair in coordinates:
    our_data[pair[0]][pair[1]] = np.nan
print(our_data)
print(np.sum(np.isnan(our_data)))


```

```python
#5
print(np.sum(np.isnan(our_data[:, 0])))
```

```python
#6
filtered_list = []
for line in our_data:
    if line[2]>1.5 and line[0]<5:
        filtered_list.append(line)
filtered_data = np.array(filtered_list)
print(filtered_data)
```

```python
#7
filtered_data = np.nan_to_num(filtered_data)
print(filtered_data)
```

```python
#8
print(np.unique(filtered_data, return_counts=True))
```

```python
#9
first_part = filtered_data[:len(filtered_data)//2]
second_part = filtered_data[len(filtered_data)//2:]
print(first_part)
print(second_part)
```

```python
#10
first_part = first_part[first_part[:, 0].argsort()]
print(first_part)
second_part = second_part[second_part[:, 0].argsort()[::-1]]
print(second_part)
```

```python
#11
united_data = np.vstack((first_part, second_part))
print(united_data)
```

```python
#12
flatten_data = united_data.flatten()
unique_numbers, counts = np.unique(flatten_data, return_counts=True)
print(unique_numbers[np.argmax(counts)])
```

```python
#13,14
def mult_below_adjusted_mean(col):
    print(np.mean(col))
    for idx, element in enumerate(col):
        if element < np.mean(col):
            element *= 2
        else:
            element /= 4
        col[idx] = element


mult_below_adjusted_mean(united_data[:, 2])
print(united_data)
```
